```python
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
#Splitting independent and dependent variables
X = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

```python
from imblearn.over_sampling import ADASYN
adasyn = ADASYN(random_state=42)
X, y = adasyn.fit_resample(X, y)
```

```python
len(X)
```

```python
#Splitting data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
#Fitting training data to the model
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
#Model accuracy
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
```

```python
#Fitting training data to the model
from sklearn.tree import DecisionTreeClassifier
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
#Model accuracy
dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)
```

```python
#Fitting K-NN classifier to the training set
from sklearn.neighbors import KNeighborsClassifier
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
knn_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_knn_pred= knn_model.predict(X_test)
y_knn_pred
```

```python
#Model accuracy
knn_cr=classification_report(y_test, y_knn_pred)
print(knn_cr)
```

```python
#Fitting Gaussian Naive Bayes classifier to the training set
from sklearn.naive_bayes import GaussianNB
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_gnb_pred= gnb_model.predict(X_test)
y_gnb_pred
```

```python
#Model accuracy
gnb_cr=classification_report(y_test, y_gnb_pred)
print(gnb_cr)
```

```python
#Fitting Multinomial Naive Bayes classifier to the training set
from sklearn.naive_bayes import MultinomialNB
mnb_model = MultinomialNB()
mnb_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_mnb_pred= mnb_model.predict(X_test)
y_mnb_pred
```

```python
#Model accuracy
mnb_cr=classification_report(y_test, y_mnb_pred)
print(mnb_cr)
```

```python
#Fitting SVC to the training set
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
# K-Fold Cross Validation

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

k = 10
kf = KFold(n_splits=k, shuffle=True, random_state=42)


# Logistic regerssion model
lr_model_scores = cross_val_score(lr_model,X, y, cv=kf)

# Decision tree model
dt_model_scores = cross_val_score(dt_model,X, y, cv=kf)

# KNN model
knn_model_scores = cross_val_score(knn_model,X, y, cv=kf)

# Gaussian naive bayes model
gnb_model_scores = cross_val_score(gnb_model,X, y, cv=kf)

# Multinomial naive bayes model
mnb_model_scores = cross_val_score(mnb_model,X, y, cv=kf)

# Support Vector Classifier model
svc_model_scores = cross_val_score(svc_model,X, y, cv=kf)

# Random forest model
rf_model_scores = cross_val_score(rf_model,X, y, cv=kf)


# Multi-layer perceptron model
mlp_model_scores = cross_val_score(mlp_model,X, y, cv=kf)

# Gradient boost model
gb_model_scores = cross_val_score(gb_model,X, y, cv=kf)


print("Logistic regression models' average accuracy:", np.mean(lr_model_scores))
print("Decision tree models' average accuracy:", np.mean(dt_model_scores))
print("KNN models' average accuracy:", np.mean(knn_model_scores))
print("Gaussian naive bayes models' average accuracy:", np.mean(gnb_model_scores))
print("Multinomial naive bayes models' average accuracy:", np.mean(mnb_model_scores))
print("Support Vector Classifier models' average accuracy:", np.mean(svc_model_scores))
print("Random forest models' average accuracy:", np.mean(rf_model_scores))
print("Multi-layer perceptron models' average accuracy:", np.mean(mlp_model_scores))
print("Gradient boost models' average accuracy:", np.mean(gb_model_scores))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define data for all classifiers
data = {
    "Classifier": ["Logistic Regression", "Decision Tree", "KNN", "Gaussian Naive Bayes",
                   "Multinomial Naive Bayes", "SVC", "Random Forest", "MLP", "Gradient Boosting"],
    "Precision": [0.98, 0.94, 0.96, 0.92, 0.81, 0.98, 0.98, 0.97, 0.98],
    "Recall": [0.97, 0.94, 0.96, 0.92, 0.81, 0.98, 0.98, 0.97, 0.98],
    "F1-Score": [0.97, 0.94, 0.96, 0.92, 0.81, 0.98, 0.98, 0.97, 0.98],
    "Support": [120, 120, 120, 120, 120, 120, 120, 120, 120],
    "Accuracy": [0.9309, 0.9540, 0.9331, 0.8846, 0.7422, 0.9497, 0.9519, 0.9373, 0.9561]
}

# Create DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to convert it to long format for visualization
df_melted = df.melt(id_vars=["Classifier"], var_name="Metric", value_name="Score")

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x="Classifier", y="Score", hue="Metric", data=df_melted, palette="viridis")
plt.title('Performance Metrics for Different Classifiers')
plt.xlabel('Classifier')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title="Metrics", loc="upper right")

# Adjust the scale of the y-axis to better visualize all metrics
plt.ylim(0, 1.1)  # Set the y-axis limit to range from 0 to 1.1

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
from tabulate import tabulate

# Sample prediction values (replace these with your actual predictions)
y_lr_pred = np.array([1, 2, 3, 4, 5])
y_dt_pred = np.array([2, 3, 4, 5, 6])
y_knn_pred = np.array([3, 4, 5, 6, 7])
y_gnb_pred = np.array([4, 5, 6, 7, 8])
y_mnb_pred = np.array([5, 6, 7, 8, 9])
y_svc_pred = np.array([6, 7, 8, 9, 10])
y_rf_pred = np.array([7, 8, 9, 10, 11])
y_mlp_pred = np.array([8, 9, 10, 11, 12])
y_gb_pred = np.array([9, 10, 11, 12, 13])
y_test = np.array([10, 11, 12, 13, 14])

# Calculate residuals for each model
lr_residuals = y_test - y_lr_pred
dt_residuals = y_test - y_dt_pred
knn_residuals = y_test - y_knn_pred
gnb_residuals = y_test - y_gnb_pred
mnb_residuals = y_test - y_mnb_pred
svc_residuals = y_test - y_svc_pred
rf_residuals = y_test - y_rf_pred
mlp_residuals = y_test - y_mlp_pred
gb_residuals = y_test - y_gb_pred

# Calculate error metrics for each model
mae = [mean_absolute_error(y_test, y_pred) for y_pred in [y_lr_pred, y_dt_pred, y_knn_pred, y_gnb_pred, y_mnb_pred, y_svc_pred, y_rf_pred, y_mlp_pred, y_gb_pred]]
mape = [np.mean(np.abs(residuals / y_test)) * 100 for residuals in [lr_residuals, dt_residuals, knn_residuals, gnb_residuals, mnb_residuals, svc_residuals, rf_residuals, mlp_residuals, gb_residuals]]
rmse = [np.sqrt(mean_squared_error(y_test, y_pred)) for y_pred in [y_lr_pred, y_dt_pred, y_knn_pred, y_gnb_pred, y_mnb_pred, y_svc_pred, y_rf_pred, y_mlp_pred, y_gb_pred]]

# Create a DataFrame to display error metrics
error_metrics_df = pd.DataFrame({
    "Model": ["Logistic Regression", "Decision Tree", "KNN", "Gaussian Naive Bayes", "Multinomial Naive Bayes", "SVC", "Random Forest", "MLP", "Gradient Boosting"],
    "MAE": mae,
    "MAPE (%)": mape,
    "RMSE": rmse
})

# Print the error metrics table using tabulate with fancy_grid format
print(tabulate(error_metrics_df, headers='keys', tablefmt='fancy_grid'))
```
